```python
def get_std(window_size, data):
    data_X = data['Ax']
    data_Y = data['Ay']
    data_Z = data['Az']
    time = []
    for t in data['Time']:
        time.append(int(t.split('-')[1]))
    std_X = pd.Series(data_X).rolling(window_size).std()
    std_Y = pd.Series(data_Y).rolling(window_size).std()
    std_Z = pd.Series(data_Z).rolling(window_size).std()
    naReplace_X = np.std(data_X[:window_size-1])
    naReplace_Y = np.std(data_Y[:window_size-1])
    naReplace_Z = np.std(data_Z[:window_size-1])
    for i in range(window_size-1):
        std_X[i] = naReplace_X
        std_Y[i] = naReplace_Y
        std_Z[i] = naReplace_Z
    std_S = std_X + std_Y + std_Z
    return time, std_S
```

```python
def get_realtime(data):
    dic = {}
    for i in range(len(data['Time'])):
        stamp = int(data['Time'][i].split('-')[1])
        hit = data['Hit'][i] * 22
        if (math.isnan(data['DistracterPosX'][i]) or math.isnan(data['DistracterPosX'][i]) or math.isnan(data['DistracterPosX'][i])):  
            dic[stamp] = (0, hit)
        else:
            dic[stamp] = (20, hit)
    return dic
    
```

```python
def preprocess(time_out, dic):
    distractor_out = []
    hit_out = []
    for t in time_out:
        if t in dic:
            distractor_out.append(dic[t][0])
            hit_out.append(dic[t][1])
        else:
            distractor_out.append(0)
            hit_out.append(0)
    return distractor_out, hit_out            
    
```

```python
def visualization(data, time, window, hand, adhd, name):
    ori_time, std_S = get_std(window, data)
    dic = get_realtime(time)
    x_axis = []
    if bool(adhd):
        if bool(hand):
            title = "ADHD HAND STD " + name
            save = "./STD/ADHD_HAND/ADHD_HAND_STD_"+ name + ".png"
        else:
            title = "ADHD FOOT STD " + name
            save = "./STD/ADHD_FOOT/ADHD_FOOT_STD_"+ name + ".png"
    else:
        if bool(hand):
            title = "HEALTHY HAND STD " + name
            save = "./STD/HEALTHY_HAND/HEALTHY_HAND_STD "+ name + ".png"
        else:
            title = "HEALTHY FOOT STD " + name
            save = "./STD/HEALTHY_FOOT/HEALTHY_FOOT_STD "+ name + ".png"       
    distractor, hit = preprocess(ori_time, dic)
    
    plt.figure(figsize=(20,5))
    plt.title(title)
    plt.xlabel("time unit:" + str(window) + "s")
    plt.ylabel("std")
    plt.ylim((0,30))
    
    plt.plot(ori_time, distractor, "x",color="red", markersize=4)
    plt.xticks(np.arange(min(ori_time), max(ori_time)+1, window))
    locs, labels = plt.xticks()
    plt.xticks(locs, range(len(ori_time)//window))
    plt.plot(ori_time, hit, ".", color="green", markersize = 10)
    plt.plot(ori_time, std_S)
    
    plt.savefig(save)
    plt.show()
```
