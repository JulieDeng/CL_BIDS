```python
import numpy as np
#from numpy import linspace
import matplotlib.pyplot as plt
from scipy import fft, arange

def plotSpectrum(y,Fs):
    n = len(y) # length of the signal
    k = arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(round(n/2))] # one side frequency range

    Y = fft(y)/n # fft computing and normalization
    Y = Y[range(round(n/2))]

    plt.plot(frq,np.absolute(Y),'r') # plotting the spectrum
    plt.xlabel('Freq (Hz)')
    plt.ylabel('|Y(freq)|')

Fs = 20.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval


#subplot(2,1,2)
plotSpectrum(adhd3['Ax'],Fs)
plt.show()
```

```python
def PowerSpectrumCumulative(y,Fs):
    n = len(y) # length of the signal
    k = arange(n)
    T = n/Fs
    frq = k/T # two sides frequency range
    frq = frq[range(round(n/2))] # one side frequency range

    Y = fft(y)/n # fft computing and normalization
    Y = Y[range(round(n/2))]
    Z = []
    for i in range(len(Y)):
        Z.append(np.absolute(Y[i])**2)
    
    for i in range(len(Z)):
        if i == 0:
            pass
        else:
            Z[i] = Z[i] + Z[i-1]
        
    plt.plot(frq, Z/Z[len(Z)-1],'r') # plotting the spectrum
    plt.xlabel('Freq (Hz)')
    plt.ylabel('|Y(freq)|')

Fs = 20.0;  # sampling rate
Ts = 1.0/Fs; # sampling interval


#subplot(2,1,2)
PowerSpectrumCumulative(adhd3['Az'],Fs)
plt.show()
```

```python
from scipy.signal import butter, lfilter


def butter_lowpass(lowcut, fs, order):
    nyq = 0.5 * fs
    low = lowcut / nyq
    b, a = butter(order, [low], btype='lowpass')
    return b, a


def butter_lowpass_filter(data, lowcut, fs, order):
    b, a = butter_lowpass(lowcut, fs, order=order)
    y = lfilter(b, a, data)
    return y


if __name__ == "__main__":
    import numpy as np
    import matplotlib.pyplot as plt
    from scipy.signal import freqz

    # Sample rate and desired cutoff frequencies (in Hz).
    fs = 20.0
    lowcut = 3.0

    # Plot the frequency response for a few different orders.
    plt.figure(1)
    plt.clf()
    for order in [3, 6, 9]:
        b, a = butter_lowpass(lowcut, fs, order=order)
        w, h = freqz(b, a, worN=2000)
        plt.plot((fs * 0.5 / np.pi) * w, abs(h), label="order = %d" % order)

    plt.plot([0, 0.5 * fs], [np.sqrt(0.5), np.sqrt(0.5)],
             '--', label='sqrt(0.5)')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Gain')
    plt.grid(True)
    plt.legend(loc='best')
    plt.show()

    # Filter a noisy signal.
    T = 14809/fs
    nsamples = 14809
    t = np.linspace(0, T, nsamples, endpoint=False)
    plt.figure(1)
    plt.clf()
    plt.plot(t, adhd['Ax'], label='Noisy signal')
    plt.xlabel('time (secs)')
    plt.hlines([-1, 1], 0, T, linestyles='--')
    plt.grid(True)
    plt.axis('tight')
    plt.legend(loc='upper left')
    plt.show()

    y = butter_lowpass_filter(adhd['Ax'], lowcut, fs, order=5)
    plt.figure(1)
    plt.clf()
    plt.plot(t, y, label='Filtered signal (%g Hz)' % lowcut)
    plt.xlabel('time (secs)')
    plt.hlines([-1, 1], 0, T, linestyles='--')
    plt.grid(True)
    plt.axis('tight')
    plt.legend(loc='upper left')
    plt.show()
```
