```python
def get_realtime(data):
    dic = {}
    for i in range(len(data['Time'])):
        stamp = int(data['Time'][i].split('-')[1])
        hit = data['Hit'][i] * 97
        if (math.isnan(data['DistracterPosX'][i]) or math.isnan(data['DistracterPosX'][i]) or math.isnan(data['DistracterPosX'][i])):  
            dic[stamp] = (0, hit)
        else:
            dic[stamp] = (90, hit)
    return dic
```

```python
def get_displacement(data, name, hand, adhd):
    
    diff = data.set_index('Time').diff().reset_index()
    total_diff = diff["Ax"] + diff["Ay"] + diff["Az"]
    pytho = np.sqrt(diff["Ax"]**2 + diff["Ay"]**2 + diff["Az"]**2)
    time = []
    for t in diff['Time']:
        time.append(int(t.split('-')[1]))
    return pytho, time
```

```python
def preprocess(time_out, dic):
    distractor_out = []
    hit_out = []
    for t in time_out:
        if t in dic:
            distractor_out.append(dic[t][0])
            hit_out.append(dic[t][1])
        else:
            distractor_out.append(0)
            hit_out.append(0)
    return distractor_out, hit_out 
```

```python
def visualization(data,time,name,adhd,hand,threshold, window):
    disp, ori_time = get_displacement(data, name, hand, adhd)
    dic = get_realtime(time)
    
    if bool(adhd):
        if bool(hand):
            title = "ADHD HAND TAT " + name
            save = "./TAT/ADHD_HAND/ADHD_HAND_TAT_"+ name + "_" + str(threshold)+ ".png"
        else:
            title = "ADHD FOOT TAT " + name
            save = "./TAT/ADHD_FOOT/ADHD_FOOT_TAT_"+ name + "_" + str(threshold)+ ".png"
    else:
        if bool(hand):
            title = "HEALTHY HAND TAT " + name
            save = "./TAT/HEALTHY_HAND/HEALTHY_HAND_TAT "+ name + "_" + str(threshold)+ ".png"
        else:
            title = "HEALTHY FOOT TAT " + name
            save = "./TAT/HEALTHY_FOOT/HEALTHY_FOOT_TAT "+ name + "_" + str(threshold)+ ".png"       
    distractor, hit = preprocess(ori_time, dic)
    TAT = len(disp[(disp>threshold)])
    percentage = TAT/len(disp) * 100
    result = title + ": "+ str(percentage) + "%\n" 
    
    plt.figure(figsize=(20,5))
    plt.title(title)
    plt.xlabel("time unit:" + str(window) + "s")
    plt.ylabel("displacement")
    plt.ylim((0,100))
    
    plt.plot(ori_time, distractor, "x",color="red", markersize=4)
    
    plt.xticks(np.arange(min(ori_time), max(ori_time)+1, window))
    locs, labels = plt.xticks()
    plt.xticks(locs, range(len(ori_time)//window))
    plt.plot(ori_time, hit, ".", color="green", markersize = 10)
    plt.plot(ori_time, disp)
    plt.axhline(y = threshold, linewidth=2, color='r')
    
    plt.savefig(save)
    plt.show()
    return result
```
